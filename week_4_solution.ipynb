## Question 2:  simulated dataset for healthy and diseased corals

```r

coral_pop <- c("Montipora capitata",
              "Porites compressa",
              "Porites lobata")

coral_pop_sample<- sample(c("Montipora capitata",
                            "Porites compressa",
                            "Porites lobata"), 100, replace=TRUE)

health_statuts <- sample(c(TRUE,FALSE), 100, replace=TRUE, prob= c(0.8,0.3))
coral_health_df <- data.frame (coral_pop_sample, health_statuts)
coral_health_tbl = as_tibble(coral_health_df)
coral_health_tbl
```

## Quesiton 3: Create a new column in coral_health_tbl called coral_cover

```r
## Question 3:
```

```r
######################
#### Exo 3############
######################

#create function for coral_cover values
coral_cover_f <- function (x){     
  
    if (coral_health_tbl$health_statuts== TRUE){
    rnorm(n=1, mean=9, sd=2)
}else{
  rnorm(n=1 ,mean=1, sd=3)}
}

#map the function to the healht status data
coral_cover <- mapply(coral_cover_f,coral_health_tbl$health_statuts)


#Add the coral_cover values to another column of the tible
coral_health_tbl_1<- coral_health_tbl %>%
  mutate(coral_cover = coral_cover)

coral_health_tbl_1
```

## Exo 4

```r
######################
#### Exo 4############
######################


binary_health_status <-sapply(coral_health_tbl_1$health_statuts,as.integer)

coral_health_tbl_2<-coral_health_tbl_1 %>%
  mutate(binary_health_status= binary_health_status)

coral_health_tbl_2
```
